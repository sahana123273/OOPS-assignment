1. What is Object-Oriented Programming (OOP)?

->Object-Oriented Programming (OOP) is a programming paradigm based on the concept of "objects", which can contain data and code. The key principles of OOP are:

    Encapsulation: Bundling data (attributes) and methods (functions) that operate on the data within a single unit (object).
    Abstraction: Hiding complex implementation details and exposing only the necessary features.

    Inheritance: Creating new classes (blueprints for objects) based on existing classes, allowing for code reusability.

    Polymorphism: Allowing objects of different classes to respond to the same method call in their own way.


2. What is a class in OOP?

->In Object-Oriented Programming (OOP), a class is a blueprint or a template for creating objects. It defines the properties (attributes) and behaviors (methods) that objects of that class will have. Think of it like a cookie cutter – the class is the cookie cutter, and the objects are the individual cookies you make with it. Each cookie (object) will have the same shape (properties) and can be decorated (methods) in similar ways, but they are distinct instances.


3. What is an object in OOP?

->In Object-Oriented Programming (OOP), an object is an instance of a class. It is a concrete realization of the blueprint defined by the class. Objects have the attributes and methods defined by their class. For example, if you have a class called Car, an object of that class would be a specific car, like a Toyota Camry with a specific color and model year.


4. What is the difference between abstraction and encapsulation?

->In Object-Oriented Programming, both abstraction and encapsulation are important principles, but they address different concerns:

    Encapsulation is about bundling data (attributes) and the methods (functions) that operate on that data within a single unit, the object. It's about keeping related things together and controlling access to the data from outside the object. Think of it like a pill capsule – the medication (data) is inside, and the capsule (object) protects it and provides a way to take it (methods).

    Abstraction is about hiding the complex implementation details and showing only the essential features of an object. It's about simplifying the view for the user of the object. Think of driving a car – you use the steering wheel, pedals, and gear shift (essential features) without needing to know the intricate details of how the engine, transmission, and braking systems work (hidden complexity).

8. How is encapsulation achieved in Python?

->In Python, encapsulation is typically achieved through a combination of conventions and mechanisms:

    Bundling Data and Methods: As mentioned before, you define attributes (data) and methods (functions) within a class. This inherently bundles related data and behavior together in an object.

    Naming Conventions (for controlling access): Python doesn't have strict access modifiers like public, private, or protected in some other languages. Instead, it uses naming conventions to indicate how attributes and methods should be accessed:
        Public: Attributes and methods without a leading underscore are considered public and can be accessed directly from outside the class.
        Protected: Attributes and methods with a single leading underscore (e.g., _attribute_name) are conventionally considered "protected." This is a hint to other developers that these members are intended for internal use within the class or its subclasses, but they can still be accessed from outside if necessary.
        
        Private: Attributes and methods with a double leading underscore (e.g., __attribute_name) are subject to name mangling. This means the Python interpreter internally changes the name of the attribute to include the class name (e.g., _ClassName__attribute_name). This makes it harder, but not impossible, to access these members directly from outside the class, providing a stronger form of encapsulation.

    Getter and Setter Methods: To provide more controlled access to attributes, you can define getter methods (to retrieve attribute values) and setter methods (to modify attribute values). This allows you to add validation or other logic when attributes are accessed or modified. The @property decorator is often used in Python to make getter and setter methods look like attribute access.

13. What is a property decorator in Python?

->In Python, the property decorator is a built-in decorator that provides a convenient way to define getter, setter, and deleter methods for class attributes. It allows you to access and modify attributes as if they were public instance variables, while still controlling how those attributes are accessed or modified behind the scenes.

Essentially, the property decorator turns a method into a "getter" for an attribute. You can then define additional methods with the same name as the property, decorated with @<property_name>.setter and @<property_name>.deleter, to handle setting and deleting the attribute, respectively.

class MyClass:
    def __init__(self, value):
        self._value = value # Use a private attribute to store the actual value

    @property
    def value(self):
        """The getter for the 'value' property."""
        print("Getting value...")
        return self._value

To add a setter, you define another method with the same name as the property and decorate it with @<property_name>.setter:

class MyClass:
    def __init__(self, value):
        self._value = value

    @property
    def value(self):
        """The getter for the 'value' property."""
        print("Getting value...")
        return self._value

    @value.setter
    def value(self, new_value):
        """The setter for the 'value' property with validation."""
        print(f"Setting value to {new_value}...")
        if new_value < 0:
            raise ValueError("Value cannot be negative")
        self._value = new_value


14. Why is polymorphism important in OOP?


->Polymorphism is a cornerstone of Object-Oriented Programming (OOP) and brings several significant benefits:

    Flexibility and Extensibility: Polymorphism allows you to write code that can work with objects of different classes that share a common interface (methods with the same name). This makes your code more flexible and easier to extend. You can add new classes that adhere to the same interface without having to modify the existing code that uses that interface.
        Example: Imagine you have a function that processes different types of shapes (circles, squares, triangles). Without polymorphism, you might need separate if-elif-else statements or switch cases to handle each shape type. With polymorphism, if each shape class has a calculate_area() method, you can call calculate_area() on any shape object, and the correct implementation for that specific shape will be executed. If you add a new shape type, you just need to implement the calculate_area() method in the new class, and the existing processing function will automatically work with it.

    Code Reusability: By allowing objects of different classes to be treated uniformly, polymorphism promotes code reusability. You can write generic code that operates on a base class or interface, and this code will work with any subclass that implements that interface.

    Simplified Code: Polymorphism simplifies your code by reducing the need for conditional logic to handle different object types. This makes your code cleaner, more readable, and easier to maintain.

    Dynamic Binding: Polymorphism relies on dynamic binding (also known as late binding). This means that the specific method to be executed is determined at runtime, based on the actual type of the object, rather than at compile time. This dynamic behavior is what makes polymorphism so powerful for handling variations in object types.

    Improved Design: Polymorphism encourages better object-oriented design by promoting the idea of programming to an interface rather than an implementation. This leads to more loosely coupled and modular code.


15. What is an abstract class in Python?

->In Python, an abstract class is a class that cannot be instantiated directly. It's designed to be a blueprint for other classes, providing a common interface and potentially some base functionality that subclasses must implement. Abstract classes are used to define a set of methods that must be implemented by any non-abstract subclass.

 The key characteristics of abstract classes in Python:

    Cannot be Instantiated: You cannot create an object directly from an abstract class. Attempting to do so will raise a TypeError.

    Contain Abstract Methods: Abstract classes can contain one or more abstract methods. An abstract method is a method declared in the abstract class but without an implementation. Subclasses are required to provide an implementation for all abstract methods.

    Can Contain Concrete Methods: Abstract classes can also contain concrete (regular) methods with implementations. These methods can be inherited and used by subclasses.

    Defined using abc module: Python's built-in abc (Abstract Base Classes) module provides the necessary tools to create abstract classes and abstract methods. You inherit from ABC to create an abstract class and use the @abstractmethod decorator for abstract methods.

17. What is the difference between a class variable and an instance variable?

->Instance Variables:

    Definition: Instance variables are specific to each individual object (instance) of a class. They are defined within the methods of a class, typically in the __init__ constructor using self.<variable_name>.
    Scope: Each instance has its own copy of the instance variables. Changes made to an instance variable in one object do not affect the instance variable of other objects.
    Purpose: They are used to store data that is unique to each instance of the class.

Class Variables:

    Definition: Class variables are shared among all instances of a class. They are defined directly within the class definition, outside of any method.
    Scope: There is only one copy of a class variable, and all instances of the class access and share this same copy. Changes made to a class variable are reflected in all instances.
    Purpose: They are used to store data that is common to all instances of the class, or to define constants that are related to the class.

     Example to illustrate the difference:

class Dog:
    # Class variable: Shared among all instances
    species = "Canis familiaris"

    def __init__(self, name, breed):
        # Instance variables: Unique to each instance
        self.name = name
        self.breed = breed

    def bark(self):
        print(f"{self.name} says Woof!")

# Creating instances of the Dog class
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Lucy", "Labrador")

# Accessing instance variables (unique to each instance)
print(f"{dog1.name} is a {dog1.breed}")
print(f"{dog2.name} is a {dog2.breed}")

# Accessing the class variable (shared by all instances)
print(f"All dogs are of the species: {Dog.species}")
print(f"{dog1.name}'s species is {dog1.species}")
print(f"{dog2.name}'s species is {dog2.species}")

# Modifying an instance variable (only affects that instance)
dog1.breed = "Poodle"
print(f"{dog1.name} is now a {dog1.breed}")
print(f"{dog2.name} is still a {dog2.breed}") # Lucy's breed is unchanged

# Modifying a class variable (affects all instances)
Dog.species = "Canis lupus familiaris"
print(f"Now all dogs are of the species: {Dog.species}")
print(f"{dog1.name}'s species is now {dog1.species}")
print(f"{dog2.name}'s species is now {dog2.species}")

19. Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python?

->The __str__ and __repr__ methods, often referred to as "dunder" methods (due to the double underscores), are special methods in Python that define how objects of your custom classes are represented as strings. While they both serve the purpose of providing a string representation, they have different intended audiences and use cases.

__str__ (for the user)

    Purpose: The __str__ method is intended to return a "user-friendly" string representation of an object. This string is meant to be readable and understandable to humans.
    When it's called: It's called by the built-in str() function and by the print() function.
    Output: The output of __str__ should be concise and informative, focusing on the essential details that a user would want to see.

__repr__ (for the developer)

    Purpose: The __repr__ method (short for "representation") is intended to return a string representation of an object that is unambiguous and can be used to recreate the object (if possible). This string is primarily for developers, useful for debugging and introspection.
    When it's called: It's called by the built-in repr() function and in interactive Python sessions (like the Colab notebook) when you simply type the variable name and press Enter. If __str__ is not defined for a class, and str() or print() are called, Python will fall back to calling __repr__.
    Output: The output of __repr__ should be as unambiguous as possible and ideally include enough information to recreate the object. A common convention is to return a string that looks like a valid Python expression to recreate the object.


Example:

class MyClass:
    def __init__(self, x, y):
        self.x = x

    def __repr__(self):
        """Developer-friendly representation."""
        return f"MyClass(x={self.x}, y={self.y})"

What is method overriding in OOP?

Method overriding in Object-Oriented Programming (OOP) is a feature that allows a subclass to provide a specific implementation of a method that is already defined in its superclass. When a method is overridden, the version of the method in the subclass is executed instead of the version in the superclass when called on an object of the subclass.

Here's how it works:

    Inheritance: You have a superclass (parent class) and a subclass (child class) that inherits from the superclass.
    Same Method Signature: Both the superclass and the subclass have a method with the exact same name and parameters.
    Different Implementation: The implementation of the method in the subclass is different from the implementation in the superclass.

When you call this method on an object of the subclass, Python's method resolution order (MRO) will first look for the method in the subclass. If it finds the method there, it executes the subclass's version. If it doesn't find the method in the subclass, it will look up the inheritance hierarchy in the superclass.

Method overriding is useful for several reasons:

    Customization: Subclasses can customize the behavior of inherited methods to suit their specific needs.
    Polymorphism: It enables polymorphism, allowing objects of different classes to respond to the same method call in their own way.
    Extensibility: It allows you to extend the functionality of existing classes without modifying them directly.

Here's an example in Python:

class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):  # Method overriding
        print("Dog barks")

class Cat(Animal):
    def speak(self):  # Method overriding
        print("Cat meows")

# Creating objects
animal = Animal()
dog = Dog()
cat = Cat()

# Calling the speak method
animal.speak()  # Output: Animal makes a sound
dog.speak()     # Output: Dog barks (subclass implementation)
cat.speak()     # Output: Cat meows (subclass implementation)

In this example, both Dog and Cat classes override the speak method from the Animal class. When speak() is called on a Dog object, the Dog class's speak method is executed. When it's called on a Cat object, the Cat class's speak method is executed. This demonstrates how method overriding allows subclasses to provide their own specific behavior for a method inherited from a superclass.

What is a property decorator in Python?

In Python, the @property decorator is a built-in decorator that provides a convenient way to define getter, setter, and deleter methods for class attributes. It allows you to access and modify attributes as if they were public instance variables, while still controlling how those attributes are accessed or modified behind the scenes.

Essentially, the @property decorator turns a method into a "getter" for an attribute. You can then define additional methods with the same name as the property, decorated with @<property_name>.setter and @<property_name>.deleter, to handle setting and deleting the attribute, respectively.

Here's a breakdown of why and how you'd use it:

Why use @property?

    Encapsulation and Control: It helps enforce encapsulation by allowing you to control how an attribute's value is accessed or changed. You can add validation, logging, or other logic within the getter and setter methods.
    Maintainability: If you later need to add validation or other logic to an attribute access, you can do so without changing the code that uses the attribute.
    Cleaner Code: It makes your code cleaner and more readable by allowing you to access methods like attributes.

How to use @property:

You define a method with the same name you want for your "property" and decorate it with @property. This method will be the getter.

class MyClass:

What is an abstract class in Python?

In Python, an abstract class is a class that cannot be instantiated directly. It's designed to be a blueprint for other classes, providing a common interface and potentially some base functionality that subclasses must implement. Abstract classes are used to define a set of methods that must be implemented by any non-abstract subclass.

Here are the key characteristics of abstract classes in Python:

    Cannot be Instantiated: You cannot create an object directly from an abstract class. Attempting to do so will raise a TypeError.
    Contain Abstract Methods: Abstract classes can contain one or more abstract methods. An abstract method is a method declared in the abstract class but without an implementation. Subclasses are required to provide an implementation for all abstract methods.
    Can Contain Concrete Methods: Abstract classes can also contain concrete (regular) methods with implementations. These methods can be inherited and used by subclasses.
    Defined using abc module: Python's built-in abc (Abstract Base Classes) module provides the necessary tools to create abstract classes and abstract methods. You inherit from ABC to create an abstract class and use the @abstractmethod decorator for abstract methods.

Why use Abstract Classes?

    Enforce Interface: Abstract classes define a contract or interface that subclasses must adhere to. This ensures that all subclasses have a consistent set of methods, which is important for maintaining code structure and predictability.
    Code Reusability: Abstract classes can provide common methods and attributes that are shared among subclasses, promoting code reusability.
    Framework Design: They are often used in framework design to define the basic structure and required components that developers must implement when extending the framework.

Example:

from abc import ABC, abstractmethod

class Shape(ABC):  # Inherit from ABC to make it an abstract class
    @abstractmethod
    def calculate_area(self):
        """Abstract method to calculate the area of the shape."""
        pass # Abstract methods typically have no implementation

    @abstractmethod
    def calculate_perimeter(self):
        """Abstract method to calculate the perimeter of the shape."""
        pass

    def display_info(self):
        """Concrete method to display information about the shape."""
        print("This is a shape.")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        """Implementation for calculating the area of a circle."""
        return 3.14 * self.radius ** 2

    def calculate_perimeter(self):
        """Implementation for calculating the perimeter of a circle."""
        return 2 * 3.14 * self.radius

class Square(Shape):
    def __init__(self, side_length):
        self.side_length = side_length

    def calculate_area(self):
        """Implementation for calculating the area of a square."""
        return self.side_length ** 2

    def calculate_perimeter(self):
        """Implementation for calculating the perimeter of a square."""
        return 4 * self.side_length

What is the difference between a class variable and an instance variable?

Great question! Understanding the difference between class variables and instance variables is fundamental to working with classes in Python.

Here's a breakdown:

Instance Variables:

    Definition: Instance variables are specific to each individual object (instance) of a class. They are defined within the methods of a class, typically in the __init__ constructor using self.<variable_name>.
    Scope: Each instance has its own copy of the instance variables. Changes made to an instance variable in one object do not affect the instance variable of other objects.
    Purpose: They are used to store data that is unique to each instance of the class.

Class Variables:

    Definition: Class variables are shared among all instances of a class. They are defined directly within the class definition, outside of any method.
    Scope: There is only one copy of a class variable, and all instances of the class access and share this same copy. Changes made to a class variable are reflected in all instances.
    Purpose: They are used to store data that is common to all instances of the class, or to define constants that are related to the class.

Here's an example to illustrate the difference:

class Dog:
    # Class variable: Shared among all instances
    species = "Canis familiaris"

    def __init__(self, name, breed):
        # Instance variables: Unique to each instance
        self.name = name
        self.breed = breed

    def bark(self):
        print(f"{self.name} says Woof!")

# Creating instances of the Dog class
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Lucy", "Labrador")

# Accessing instance variables (unique to each instance)
print(f"{dog1.name} is a {dog1.breed}")
print(f"{dog2.name} is a {dog2.breed}")

# Accessing the class variable (shared by all instances)
print(f"All dogs are of the species: {Dog.species}")
print(f"{dog1.name}'s species is {dog1.species}")
print(f"{dog2.name}'s species is {dog2.species}")

# Modifying an instance variable (only affects that instance)
dog1.breed = "Poodle"
print(f"{dog1.name} is now a {dog1.breed}")
print(f"{dog2.name} is still a {dog2.breed}") # Lucy's breed is unchanged

# Modifying a class variable (affects all instances)
Dog.species = "Canis lupus familiaris"
print(f"Now all dogs are of the species: {Dog.species}")
print(f"{dog1.name}'s species is now {dog1.species}")
print(f"{dog2.name}'s species is now {dog2.species}")

In summary:

    Instance variables: Data specific to each object.
    Class variables: Data shared among all objects of a class.

Using the correct type of variable is important for organizing and managing the data within your classes effectively.

What is multiple inheritance in Python?

Multiple inheritance is a feature in Object-Oriented Programming where a class can inherit properties and behaviors from more than one parent class. In Python, a class can inherit from multiple base classes by listing them in the class definition's parentheses, separated by commas.

Here's the basic syntax:

class BaseClass1:
    # ...

class BaseClass2:
    # ...

class DerivedClass(BaseClass1, BaseClass2):
    # ...

In this example, DerivedClass inherits from both BaseClass1 and BaseClass2. This means that an object of DerivedClass will have access to the methods and attributes of both parent classes.

How Python handles method resolution (MRO):

When you call a method on an object of a class that uses multiple inheritance, Python needs a way to determine which parent class's method to execute, especially if the same method name exists in multiple parent classes. This is handled by the Method Resolution Order (MRO).

Python 2 used a depth-first, left-to-right MRO algorithm. Python 3 uses the C3 Linearization algorithm, which provides a more consistent and predictable MRO, especially in complex inheritance hierarchies.

You can see the MRO of a class using the .__mro__ attribute or the help() function:

class A:
    def greet(self):
        print("Hello from A")

class B(A):
    def greet(self):
        print("Hello from B")

class C(A):
    def greet(self):
        print("Hello from C")

class D(B, C):
    pass

d = D()
d.greet() # Output will depend on the MRO

print(D.__mro__)

In this case, the MRO of D would typically be (D, B, C, A, object). When d.greet() is called, Python looks for greet in D, then B, then C, then A, and finally object. The first one it finds is executed.

Potential Issues with Multiple Inheritance:

While powerful, multiple inheritance can sometimes lead to complexities, particularly with the "Diamond Problem." This occurs when a class inherits from two classes that have a common ancestor. The C3 linearization algorithm in Python 3 is designed to handle the diamond problem gracefully by ensuring that the common ancestor is not called more than once.

Advantages of Multiple Inheritance:

    Code Reusability: Allows a class to inherit useful features from multiple existing classes.
    Modeling Complex Relationships: Can be useful for modeling real-world scenarios where an entity possesses characteristics from multiple categories.

Disadvantages of Multiple Inheritance:

    Complexity: Can make the inheritance hierarchy more difficult to understand and manage.
    Method Resolution Issues: While Python 3's MRO is improved, understanding how methods are resolved in complex hierarchies can still be challenging.
    Increased Coupling: Can lead to tighter coupling between classes.

Due to the potential for complexity, it's often recommended to use multiple inheritance judiciously and consider alternative approaches like using mixins or composition when appropriate.

While they both serve the purpose of providing a string representation, they have different intended audiences and use cases.

Here's a breakdown of each:
__str__ (for the user)

    Purpose: The __str__ method is intended to return a "user-friendly" string representation of an object. This string is meant to be readable and understandable to humans.
    When it's called: It's called by the built-in str() function and by the print() function.
    Output: The output of __str__ should be concise and informative, focusing on the essential details that a user would want to see.

__repr__ (for the developer)

    Purpose: The __repr__ method (short for "representation") is intended to return a string representation of an object that is unambiguous and can be used to recreate the object (if possible). This string is primarily for developers, useful for debugging and introspection.
    When it's called: It's called by the built-in repr() function and in interactive Python sessions (like the Colab notebook) when you simply type the variable name and press Enter. If __str__ is not defined for a class, and str() or print() are called, Python will fall back to calling __repr__.
    Output: The output of __repr__ should be as unambiguous as possible and ideally include enough information to recreate the object. A common convention is to return a string that looks like a valid Python expression to recreate the object.

Key Differences and Relationship:

    Audience: __str__ is for users, __repr__ is for developers.
    Goal: __str__ aims for readability, __repr__ aims for unambiguity and reproducibility.

What is the significance of the __del__ method in Python?

purpose and significance:

    Cleanup and Resource Release: The main use case for __del__ is to clean up resources that the object has acquired during its lifetime.     Alternatives for Resource Management: Due to the uncertainty of __del__ execution, it's generally recommended to use alternative mechanisms for resource management when possible. Context managers provide a more reliable way to ensure that cleanup actions are performed, even if exceptions occur.

    Decorators with Arguments: The __call__ method is often used when creating decorators that accept arguments. The decorator itself is a class, and its __call__ method is executed when the decorated function is called.

Example:

class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, number):
        """Allows instances of Multiplier to be called like a function."""
        print(f"Multiplying {number} by {self.factor}")
        return number * self.factor

# Creating an instance of Multiplier
double = Multiplier(2)
triple = Multiplier(3)

# Calling the instances like functions
result1 = double(10)
result2 = triple(5)

print(f"Result 1: {result1}")
print(f"Result 2: {result2}")

# Example illustrating state:
another_doubler = Multiplier(2)
print(f"Another double of 7: {another_doubler(7)}")

```python
#2.  Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

from abc import ABC, abstractmethod

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Derived class for Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

# Derived class for Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Create objects
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Display areas
print("Area of Circle:", circle.area())
print("Area of Rectangle:", rectangle.area())

```

```python
#5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance

# Class demonstrating encapsulation
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance   # Private attribute

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ₹{amount}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: ₹{amount}")
        else:
            print("Insufficient balance or invalid amount.")

    # Method to check balance
    def check_balance(self):
        print(f"Current Balance: ₹{self.__balance}")

# Create a BankAccount object
account = BankAccount(1000)

# Use the methods
account.check_balance()
account.deposit(500)
account.withdraw(300)
account.check_balance()



```

```python
#12 Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # List of grades

    def average_grade(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)


# Create Student objects
s1 = Student("Alice", [85, 90, 78, 92])
s2 = Student("Bob", [70, 75, 80])

# Display average grades
print(f"{s1.name}'s average grade: {s1.average_grade():.2f}")
print(f"{s2.name}'s average grade: {s2.average_grade():.2f}")

```

```python
#14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

# Base class
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate


# Derived class
class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)  # Inherit from Employee
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()  # Get base salary from Employee
        return base_salary + self.bonus


# Create Employee and Manager objects
emp1 = Employee("Alice", 40, 200)
mgr1 = Manager("Bob", 45, 250, 5000)

# Display salaries
print(f"{emp1.name}'s Salary: ₹{emp1.calculate_salary()}")
print(f"{mgr1.name}'s Salary (with bonus): ₹{mgr1.calculate_salary()}")
```
